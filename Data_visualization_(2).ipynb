```python
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn import tree
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
df1=pd.read_csv('house3.csv')
df1.head()
```

```python

plt.figure(figsize=(15,10))
sns.scatterplot(y='price',x='total_sqft',data=df1,color='red')
```

```python

plt.figure(figsize=(15,10))
sns.lmplot(y='price',x='total_sqft',data=df1)
```

```python

plt.figure(figsize=(15,10))
sns.jointplot(y='price',x='total_sqft',data=df1,color='violet')
```

```python

plt.figure(figsize=(15,10))
sns.kdeplot(df1['total_sqft'],shade=True,color='cyan')
```

```python
plt.figure(figsize=(15,10))
sns.distplot(df1['total_sqft'],color='red')
```

```python
ax=sns.lineplot(x='total_sqft',y='price',data=df1,color='grey')
```

```python
ax=sns.lineplot(x='total_sqft',y='bhk',data=df1,color='green')
```

```python
ax=sns.lineplot(x='total_sqft',y='bath',data=df1,color='yellow')
```

```python
ax=sns.lineplot(x='price',y='bath',data=df1,color='pink')
```

```python
ax=sns.lineplot(x='price',y='bhk',data=df1,color='black')
```

# Plottting categorical data

```python
sns.catplot(x="price",y="area_type",data=df1)
```

```python
sns.catplot(x="area_type", y="bhk", kind="swarm", data=df1);
```

```python
sns.catplot(x="total_sqft",y="area_type",kind='box',data=df1)
```

```python
sns.catplot(x="price",y="area_type",kind='box',data=df1)
```

```python
sns.catplot(x="total_sqft",y="location",kind='box',data=df2)
```

```python
sns.catplot(x="total_sqft",y="location",data=df2)
```

# Statistical estimation within categories

```python
ax = sns.catplot(x='area_type',kind='count',data=df1,orient="h")
ax.fig.autofmt_xdate()
```

```python
#for only first 200 values
ax = sns.catplot(x='location',kind='count',data=df2,orient="h")
ax.fig.autofmt_xdate()
```
