```python
df1=df.dropna()
df1.shape
```

```python
df1
```

```python
dummies=pd.get_dummies(df1.Sex)
dummies.head(3)
```

```python
df2=pd.concat([df1,dummies.drop('male',axis='columns')],axis='columns')
df2.head()
```

```python
dummies=pd.get_dummies(df2.Embarked)
dummies.head(3)
```

```python
df3=pd.concat([df2,dummies.drop('S',axis='columns')],axis='columns')
df3.head()
```

```python
target=df3["Survived"].values
features=df3[['Pclass','Age','SibSp','Parch','female','Fare','C','Q']].values
```

```python
features
```

```python
from sklearn import linear_model,preprocessing
classifier= linear_model.LogisticRegression()
```

```python
cl=classifier.fit(features,target)
cl
```

```python
cl.score(features,target)
```

```python
cl.predict()
```

```python
poly= preprocessing.PolynomialFeatures(degree=2)
poly_features= poly.fit_transform(features)
```

```python
cl=classifier.fit(poly_features,target)
cl
```

```python
cl.score(poly_features,target)
```

```python
from sklearn import tree,model_selection
```

```python
target=df3["Survived"].values
features_name=['Pclass','Age','SibSp','Parch','female','Fare','C','Q']
features=df3[features_name].values
```

```python
decsion_tree=tree.DecisionTreeClassifier(random_state=1)
decsion_tree_=decsion_tree.fit(features,target)
```

```python
decsion_tree_.score(features,target)
```

```python
scores=model_selection.cross_val_score(decsion_tree,features,target,scoring='accuracy',cv=50)
scores
```

```python
scores.mean()
```

```python
generalized_tree = tree.DecisionTreeClassifier(
random_state=1,
max_depth=7,
min_samples_split=2
)
```

```python
generalized_tree_=generalized_tree.fit(features,target)
```

```python
generalized_tree_.score(features,target)
```

```python
scores=model_selection.cross_val_score(generalized_tree,features,target,scoring='accuracy',cv=50)
scores
```

```python
scores.mean()
```

```python
tree.export_graphviz(generalized_tree_,features_name=features_name,out_file="tree.dot")
```
